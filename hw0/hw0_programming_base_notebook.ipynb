```python
# Import libraries:
import numpy as np
import pandas as pd
import seaborn as sns  # optional, but recommended
import matplotlib.pyplot as plt 
from sklearn import datasets
```

```python
# load dataset
cancer = datasets.load_breast_cancer()
```

```python
# get keys from dictionary
cancer.keys()
```

```python
# print the dataset description
print(cancer.DESCR)
```

```python
# Check feature names:
cancer.feature_names
```

```python
# load data into panda dataframe
df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
# selecting column by column name
df['mean radius']
```

```python
# select multiple rows
df[:100]
```

```python
# select one row
df[99:100]
# another way to select row
# df.iloc[99]
```

#### Question 1

**How many rows and columns are there in this data set?**

```python
rows, columns = df.shape
print("Amount of rows: ", rows)
print("Amount of columns: ", columns)
```

**What do the rows and columns represent?**  
The rows represent the numbers that describe the breast mass.
The columns are the feature names.

#### Question 2

**How many malignant cases are there in total?**

```python
# 0 = malignant
dataset_targets = cancer.target
malignant_count = np.where(dataset_targets == 0)[0]
print("Number of malignant cases: ", len(malignant_count))
```

#### Question 3

**Make a scatterplot of 'mean compactness' and 'mean perimeter'. Try to color the points by their label (i.e. malignant or benign). You may find scatter (from matplotlib) or jointplot (from seaborn) useful for this.**

```python
df['diagnosis'] = cancer.target

malignant_subset = df[df['diagnosis'] == 0]
plt.scatter(malignant_subset['mean compactness'], malignant_subset['mean perimeter'], label="malignant", color="red")
benign_subset = df[df['diagnosis'] == 1]
plt.scatter(benign_subset['mean compactness'], benign_subset['mean perimeter'], label="benign", color="blue")
```

#### Question 4

**Make a pairwise scatterplot of 'mean texture', 'mean perimeter', 'mean compactness', and 'mean radius'. You may find scatter_matrix (from matplotlib) or pairplot (from seaborn) useful for this.**

```python
color_palette = {0: "red", 1: "blue"}
sns.pairplot(df, vars=['mean texture', 'mean perimeter', 'mean compactness', 'mean radius'], hue='diagnosis', palette=color_palette)
```

**Do you notice that two of these features seem to be particularly correlated?**  
The perimeter and radius
**Why might this be?**  
because perimeter affects radius geometically
